```python
# 특정 값 반환
print(df.loc['two', 'name'])
print()
# 특정 행 반환 => 시리즈
print(df.loc['two', :])
# print(df.loc['two'])
print()
# 특정 열 반환 => 시리즈
print(df.loc[: , 'name'])
# print(df['name'])
print()
# 연속적이지 않은 행 반환 => 데이타프레임 
print(df.loc[['one', 'four'], :])
print()
# 연속적이지 않은 열 반환 => 데이타프레임 
print(df.loc[: , ['name', 'points']])
# print(df[['name', 'points']])
```

```python
# 연속적인 행 반환 => 데이타프레임 
print(df.loc['three':'five', :])
print()
# 연속적인 컬럼 반환 => 데이타프레임 
print(df.loc[: , 'name':'points'])
print()
# 연속적인 특정행과 특정컬럼 반환 => 데이타프레임 
df.loc['three':'five', 'name':'points']
```

```python
# iloc[] 방식을 이용한 조회 
# 행위치인덱스와 열위치인덱스 이용 
```

```python
df
```

```python
# 특정값 반환 
print(df.iloc[3, 2])
print()
# 특정행 반환 => 시리즈
# print(df.iloc[5, :])
print(df.iloc[5])
print()
# 특정컬럼 반환 => 시리즈
print(df.iloc[:, 1])
print()
# 불연속인 특정행, 특정열 반환 => 데이타프레임 
# 4, 7 행중에서 year, penalty 컬럼 데이타만 조회 
print(df.iloc[[3, 6], [0, 3] ])
# 연속적인 행범위, 열범위 => 데이타프레임
# 3~6행 중에서 1,2열 반환 => 데이타프레임 
print()
print(df.iloc[2:6, :2])
```

```python
# 컬럼의 추가와 수정 
# df[컬럼명] = 값|리스트|컬럼을이용한수식
# 컬럼명이 있다면 데이타 값 수정
# 컬럼명이 없다면 새로운 컬럼으로 추가 
data = {'Math':[100, 45, 89, 40],
        'Eng':[80, 95, 88, 77],
        'Sci':[100, 77, 89, 100]}
df = pd.DataFrame(data, index=['김길동', '고길동', '박길동','최길동'])
df
```

```python
# Eng 컬럼값으로 모두 100 으로 수정
df['Eng'] = 100
# Math 컬럼값을 리스트를 이용해서 수정
df['Math'] = [80, 80, 30, 30]
# df.loc[:, 'Math'] = [80, 80, 30, 30]
# df.iloc[:, 0] = [80, 80, 30, 30]
df
```

```python
# Python 과목 추가 
df['Python'] = [50, 60, 70, 80]
df
```

```python
# 총점, 평균, 합격 컬럼 추가 
df['Total'] = df['Math'] + df['Eng'] + df['Sci'] + df['Python']
df['Avg'] = df['Total']/4
# 평균 80점 이상이면 True
df['Pass'] = np.where(df['Avg'] >= 80, '합격', '불합격')
df
```

# 집계함수 

```python
data = { 'kor':[40, 55, 66],
        'math':[70, 55, 66],
        'eng':[80, 95, 100]}
index=['user1','user2','user3']
df = pd.DataFrame(data, index=index)
df
```

```python
# 영어점수의 평균은?
round(df['eng'].mean(), 2)
```

```python
# user2의 총점과 평균은?
df.loc['user2'].sum(), round(df.loc['user2'].mean(), 2)
```

```python
# 컬럼별 평균
# df.mean()
df.mean(axis=0)
```

```python
# 행단위 평균
df.mean(axis=1)
```

```python
# 총점과 평균을 컬럼으로 추가하기 
data = { 'kor':[40, 55, 66],
        'math':[70, 55, 66],
        'eng':[80, 95, 100]}
index=['user1','user2','user3']
df = pd.DataFrame(data, index=index)

df['tot'] = df.sum(axis=1)
df['avg'] = round(df.loc[:, 'kor':'eng'].mean(axis=1), 2)
df
```
