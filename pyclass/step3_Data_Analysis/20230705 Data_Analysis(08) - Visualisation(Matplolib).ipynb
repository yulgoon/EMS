### plot(): 라인을 그리는 함수
- plt.xlabel(X축)
- plt.ylabel(Y축)
- plt.title(제목)
- plt.legend() : 범례 표시
- plt.figure(figsize=(w,h))
- plt.plot( X축데이타, Y축데이타, label=라벨명 )

```python
# 선그래프

# 데이터 준비
x = np.arange(-5, 6)
print('x =', x)

y = 2*x
plt.plot(x, y, label='y = 2*x')
print('y =', y)

y = abs(2*x+1) # abs(): 절댓값 함수
plt.plot(x, y, label='절댓값 y = 2*x')
print('절댓값 y =', y)

# 그래프화
# plt.figure(figsize=(16, 9))
plt.legend() # 범례 표시
plt.xlabel('X')
plt.ylabel('Y')
plt.title('y=2*x')
plt.grid() # 그리드 표시
plt.savefig('output/myTestGraph.png') # 이미지 파일로 저장
plt.show()
```

```python
plt.figure(figsize=(20,7))
x = np.linspace(-5, 6,100)
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x, y1,'g^-', label='sin')
plt.plot(x, y2, label='cos', marker='o')

plt.title('sin, cos 그래프')
plt.legend()
plt.grid()
```

```python
# 그래프화 
# plt.plot(x, y, label=그래프명)
plt.figure(figsize=(5, 5)) # 그래프 크기 설정

x = np.arange(-5, 6)
y = abs(2*x) # abs(): 절댓값 함수
plt.plot(x, y, label='y = abs(2*x)', color='r', marker='o',
         linestyle='--', linewidth='1', markersize='10')

plt.plot(x, x**2, 'k^-', label='y = x**2')

plt.title('절대값 직선 그래프')
plt.xlabel('X축')
plt.ylabel('Y축')
plt.legend() # 범례표시
plt.grid() # 그래드 표시
plt.savefig('output/graph2.png') # 이미지 파일로 저장
plt.show()
```

### Pie Plot
- plt.pie(values, labels=labelList, autopct='%digit.pointNum%Unit')
- only x datas existed

```python
# Label + Value
label_list = ['Frogs', 'Hogs', 'Dogs', 'Logs']
sizes = [15, 30, 45, 78]
print(sizes, '\ntotal:', sum(sizes), '\n')
for item in sizes:
    res = item/sum(sizes)*100
    print(round(res, 2))

plt.figure(figsize=(5,5))
plt.title('파이 그래프 백분율 표시')
plt.pie(sizes, labels=label_list, autopct='%.2f%%')
plt.show()
```

## Dataframe Graph
- 데이타프레임명.plot( kind=그래프형태, figsize=(가로, 세로), title=제목)
- 데이타프레임명[컬럼명].plot( kind=그래프형태, figsize=(가로,세로), title=제목)
- 데이타프레임명.plot( kind=그래프형태, figsize=(가로,세로), title=제목 , subplots=True, layout(행수, 열수) )
- 그래프 형태 : 'line’ , 'bar’ , 'barh’ , 'hist’ , 'box’ , 'area’ , ‘pie. 'scatter'

```python
# 데이터프레임 만들기
data = {'kor':[40, 55, 66, 77, 34],
        'math':[70, 55, 66, 20, 40],
        'eng':[80, 95, 100, 77, 90],
        'sci':[90, 45, 77, 90, 67]}

name_list = ['고길동','김희동','신윤지','이민호','박준희']

df = pd.DataFrame(data, index=name_list)
df
```

```python
# 데이타프레임 전체 그래프
df.plot(kind='bar', title='중간고사 성적', figsize=(12,5)).get_legend().remove() # get_legend().remove() => legend 안 보이게 하는 함수
plt.xticks(rotation=0)
# plt.legend(['국어','수학','영어','과학'], loc='lower right')
plt.show()
```

```python
# 한 과목만 막대그래프

plt.title('국어 성적')
df['kor'].plot(kind='bar')
plt.xticks(rotation=0)
# plt.grid()
# plt.savefig('output/국어성적 막대그래프.png')
plt.legend()
plt.show()
```

```python
# 국어, 수학 성적 막대 그래프

# plt.title('국어/수학 성적')
# colours = ['blue', 'red']
df[['kor', 'math']].plot(kind='bar', color=['green', 'orange'], title='국어/수학 성적')
plt.xticks(rotation=0)
# plt.grid()
# plt.savefig('output/국어성적 막대그래프.png')
plt.legend(['국어', '수학'], loc='upper center')
plt.show()
```

```python
# 가로막대 그래프 
df[['kor', 'math']].plot(kind='barh', title='국어와 수학성적 가로막대 그래프', color=['red', 'green'])
# plt.grid()
plt.xticks(rotation=30)
plt.savefig('output/국어수학성적가로막대그래프.png')
plt.legend(['국어', '수학'], loc='upper right')
plt.show()
```

```python
# 1행의 그래프
df.loc['고길동'].plot(kind='bar', title='고길동 성적 그래프', color=['r','g','b','y'])
plt.xticks(rotation=0)
plt.show()
```

```python
# 여러행의 그래프
df.loc[['고길동', '신윤지']].plot(kind='bar', title='고길동, 신윤지 성적 그래프', color=['r','g','b','y'])
plt.xticks(rotation=0)
plt.show()
```

```python
# 컬럼별로 분리된 후 시각화
# subplot() 방식으로 행과 열로 배치하기

# 2행 3열 레이아웃 => subplot(행/열/번째)
plt.figure(figsize=(15,8))

plt.subplot(231)
df.iloc[0].plot(kind='bar', title=df.index[0], color=['r','b','g','c']);
plt.xticks(rotation=0)

plt.subplot(232)
df.iloc[1].plot(kind='bar', title=df.index[1], color=['r','b','g','c']);
plt.xticks(rotation=0)

plt.subplot(233)
df.iloc[2].plot(kind='bar', title=df.index[2], color=['r','b','g','c']);
plt.xticks(rotation=0)

plt.subplot(234)
df.iloc[3].plot(kind='bar', title=df.index[3], color=['r','b','g','c']);
plt.xticks(rotation=0)

plt.subplot(235)
df.iloc[4].plot(kind='bar', title=df.index[4], color=['r','b','g','c']);
plt.xticks(rotation=0)

# 그래프 전체 제목 설정
plt.suptitle('학생별 중간고사 성적')

plt.tight_layout()
plt.show()
```
