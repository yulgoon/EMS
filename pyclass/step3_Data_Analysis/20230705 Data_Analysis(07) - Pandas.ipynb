## 문자열 처리
- 데이타프레임[컬럼명].str[start:end:step] # str[위치인덱스]
- 데이타프레임[컬럼명].str.문자열함수

- len()
- contains()

```python
tips = sns.load_dataset('tips')
tips

tips.sample(5)
```

```python
# sex 컬럼에서 첫글자만 분리시켜서 별도 컬럼에 저장

tips['sex'][0] # 1행

tips['sex'].str[0].sample(5)

tips['r1'] = tips['sex'].str[0]
tips.sample(5)
```

```python
# time 컬럼에서 슬라이싱
tips['time'].unique() # 'Lunch'와 'Dinner'만 있네

tips['time'].sample(5)

# time 컬럼에서 'Dinner', 'Lunch' 앞 세글자만 추출 후 컬럼에 추가
tips['time'][0:3] # 첫 3행 출력

tips['time'].str[0:3]
tips['r2'] = tips['time'].str[0:3]
tips.sample(5)
```

```python
# upper(), lower()
# day 컬럼값을 모두 대문자, 소문자로 변경 후 컬럼에 추가

tips['r3'] = tips['day'].str.upper()
tips['r4'] = tips['smoker'].str.lower()
tips
```

```python
# contains(문자열) : True, False로 반환

tips['day'].unique() # 목, 금, 토, 일 값만 있네

# day 컬럼값 안에 r을 포함하고 있다면 True 반환
tips['r5'] = tips['day'].str.contains('r')
tips.sample(5)
```

```python
# replace(old, new)

tips['smoker'].unique()

# 'smoker' 컬럼값 교체
tips['r6'] = tips['smoker'].str.replace('No', '비흡연').replace('Yes', '흡연')

tips.sample(5)
```

## 그룹화(groupby)와 집계함수
- 데이타프레임.groupby(컬럼명).컬럼명.집계함수()
- 데이타프레임.groupby([컬럼명리스트]).집계함수()
    - 집계함수가 여러개일 때, df.groupby([컬럼명리스트...]).agg([집계함수리스트....])
- 그룹객체명 = 데이타프레임.groupby(컬럼명)
- 그룹객체명.집계함수()
- 그룹객체명.agg([집계함수1, 집계함수2...])
- 집계함수의 종류
    - count, size, mean, std, min, max, sum, var

```python
df = pd.read_csv('data/gapminder.tsv', sep='\t')
df.sample(5)

# lifeExp 컬럼 전체 데이터의 평균
round(df['lifeExp'].mean(), 2)
```

```python
df['pop'].mean()

# groupby 객체
df.groupby('year')

# 수치형 컬럼의 집계함수 적용 => 데이터프레임
df.groupby('year').mean()
```

```python
# 지수 => 실수 스타일 변경
pd.options.display.float_format = '{:.2f}'.format
```

```python
# 연도별 인구 컬럼의 평균 => 시리즈 반환
df.groupby('year').pop.mean()
df.groupby('year')['pop'].mean()
```

```python
# '70 ~ '80년대의 연도별 인구 평균
type(df.groupby('year')['pop'].mean()[1972])
df.groupby('year')[['pop']].mean().loc[1972:1987]
```

```python
# groupby 컬럼이 두 개 이상일 때 => multiKey groupBy

# 대륙별 연도별 수치형 데이터 컬럼의 평균
df.groupby(['continent'])['pop'].mean() # 대륙별 컬럼의 데이터 평균

# [[]] : 겹대괄호 => DataFrame type | [] : 홑대괄호 => dict type
df.groupby(['continent', 'year'])['pop'].mean()
df.groupby(['continent', 'year'])[['pop']].mean() 
```

```python
df2 = df.groupby(['continent', 'year'])[['pop']].mean()
df2.sample(5).index # listTuple style
```

```python
df2.loc[('Americas', 1952)]
```

```python
# 특정 대륙의 연도별 인구 평균
df2.loc[('Oceania', 1952):(('Oceania'), 2007)]
```

```python
# 그룹바이 객체생성 .. 대륙별
grouped_continent = df.groupby('continent')
grouped_continent
```

```python
# 집계함수 적용
grouped_continent.max()
```

```python
# 집계함수가 여러개인 경우: agg()

# 대륙별 max, min
grouped_continent.agg([np.max, np.min])
```

```python
# 대륙별 lifeExp, pop 컬럼의 max, min
grouped_continent['lifeExp', 'pop'].agg(['max', 'min'])
```

```python
df4 = grouped_continent['lifeExp', 'pop'].agg(['max', 'min'])
df4
```

## 데이터 변경
- replace()
    - 데이타프레임명.컬럼명.replace({기존값:새로운값})
    
- map() # Series만 가능
    - 데이타프레임명.컬럼명.map({기존값:새로운값})
    - 데이타프레임명.컬럼명.map(함수명 | 람다함수)
    
- apply() # DataFrame 및 Series에 모두 적용 가능, 주로 많이 씀
    - 데이타프레임명.컬럼명.apply(함수명 | 람다함수)
    - 데이타프레임명[[컬럼명리스트]].apply(함수명 | 람다함수)
    
- applymap() # DataFrame만 가능
    - 데이타프레임명[[컬럼명리스트]].applymap(함수명 | 람다함수)

```python
# replace() : 컬럼의 특정 값 교체시 이용
'''
데이타 출처 :
https://raw.githubusercontent.com/rstudio/Intro/master/data/wages.csv
'''
# data load
df = pd.read_csv("data/wages.csv")
df.sample(5)
df.shape
df.info()
```

```python
df['race'].unique(), df['sex'].unique()
```

```python
# race 컬럼값을 정수 수치형으로 변경: replace()
df['race2'] = df['race'].replace({'white':1, 'black':2, 'hispanic':3, 'other':4})
df.sample(10)
```

```python
# race 컬럼값을 정수 수치형으로 변경: map()
df['sex2'] = df['sex'].map({'male':1, 'female':2})
df.sample(10)
```

```python
# lambda 함수와 map() 이용하여 데이터 수정: height+'inches'

# map() + 함수
def change(x):
    return f'{x} IN.'

df['height2'] = df['height'].map(change)
df
```

```python
# map() + lambda
# '$'+earn
df['earn2'] = df['earn'].map(lambda x:f'US$ {x}')
df
```

```python
df2 = pd.read_csv("data/wages.csv") # 데이터 reset and load

# 수치형 컬럼만 별도 데이터프레임으로 복사
df2 = df[['earn','height','ed','age']].copy()
df.info()
df2
```

```python
# applymap(함수명 | 람다함수) test
df3  = df2.applymap(lambda x:f'{x:.2f}')
df3
```

```python
df3 = df2[['height', 'earn']].applymap(lambda x:round(x, 1))
df3
```
