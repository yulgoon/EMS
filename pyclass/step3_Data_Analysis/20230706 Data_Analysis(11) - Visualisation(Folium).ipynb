### 특정 좌표
- folium.Map(location=[위도,경도]) => 중심위치 설정

### 서클 마커
- folium.CircleMarker([위도, 경도], radius=반지름, popup='팝업메세지', color='색상값', fill_color='색상값', ).add_to(foliumMap객체)

```python
'''
폰트 이미지, 아트, 아이콘 모음 사이트 => prefix='fa'로 지정하였기 때문에 해당 사이트의 아이콘 이름으로 사용 가능
https://fontawesome.com/icons?d=gallery
'''
```

```python
# 특정 위치의 위도, 경도 확인하기
# 위도는 37.47649364443451, 경도는 126.8799195139530

# 1) 맵 객체 생성
myMap = folium.Map(location=[37.47649364443451, 126.87991951395308], zoom_start=15, tiles='Stamen Toner') # tiles는 지도 모양

# 2) 마커 표시(추가)
folium.Marker(location=[37.47649364443451, 126.87991951395308],
              icon=folium.Icon(icon='building', color='red', prefix='fa'),
              popup='KM타워').add_to(myMap)
folium.Marker(location=[37.477909039465345, 126.88071994866505],
              icon=folium.Icon(icon='envelopes-bulk', color='red', prefix='fa'),
              popup='가산디지털 우체국').add_to(myMap)
folium.Marker(location=[37.4802607085943, 126.88530643870789],
              icon=folium.Icon(icon='building', color='red', prefix='fa'),
              popup='에코시안').add_to(myMap)
folium.Marker(location=[37.47984477651434, 126.87712441657152],
              icon=folium.Icon(icon='building', color='red', prefix='fa'),
              popup='한국EMS협회').add_to(myMap)
folium.Marker(location=[37.48161854460435, 126.88257109301044],
              icon=folium.Icon(icon='train-subway', color='darkblue', prefix='fa'),
              popup='가디역 1호선').add_to(myMap)

# 3) 서클 마커
folium.CircleMarker([37.47649364443451, 126.87991951395308],
                   radius=50, fill_color='yellow',
                   popup='KM타워').add_to(myMap)


# 4) html로 저장
myMap.save('output/가산디지털단지.html')

# 5) 콘솔 출력
myMap
```

```python
# 반복문을 활용하여 지도에 마커 찍기
'''
- police.csv => 데이타프레임 (위도, 경도, 팝업내용)
- folium 객체 생성
- 데이타프레임(위도, 경도, 팝업내용) 반복문 이용해서 마커 표시
- html로 저장
'''
import pandas as pd

df = pd.read_csv('data/police.csv', index_col=1)
df.shape
df.head()
```

```python
# Unnamed: 0 컬럼 삭제
df.columns
del df['Unnamed: 0']
```

```python
# 필요없는 unnamed 컬럼 삭제 확인
df.sample(3)
```

```python
# 위도, 경도 데이터 타입 확인
df.dtypes
```

```python
# 결측값 확인
df.isnull().sum()
```

```python
# 서울의 중앙 위도와 경도 기준
policeMap = folium.Map(location=[37.5502, 126.982], zoom_start=12, tiles='Stamen Toner')

for i in range(len(df)):
    folium.Marker([df.iloc[i,1], df.iloc[i,2]],
                 popup=df.index[i],
                 icon=folium.Icon(icon='building', color='blue',
                                 prefix='fa')).add_to(policeMap)
    
policeMap.save('output/서울경찰서지도.html')
    
policeMap
```
